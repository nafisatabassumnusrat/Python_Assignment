```python
# 1. Create a list of 5 numbers and print the sum[cite: 76].
# Hint: Use a loop with total = 0[cite: 77].
numbers = [10, 20, 30, 40, 50]
total = 0
for num in numbers:
    total += num
print("Problem 1 - Sum of numbers:", total)

# 2. Create a list of 5 fruit names and print the first and last fruit[cite: 78].
# Hint: Use index 0 and -1[cite: 79].
fruits = ["Apple", "Banana", "Mango", "Orange", "Grape"]
print(f"Problem 2 - First fruit: {fruits[0]}, Last fruit: {fruits[-1]}")

# 3. Count how many numbers in a list are greater than 10[cite: 80].
# Hint: Use a loop, if condition, and counter[cite: 81].
mixed_numbers = [5, 12, 8, 20, 3, 15]
count = 0
for num in mixed_numbers:
    if num > 10:
        count += 1
print("Problem 3 - Numbers greater than 10:", count)

# 4. Add a new item, remove an item, print list after each step[cite: 82].
# Hint: Use .append() and .remove()[cite: 83].
items = ["Pen", "Pencil", "Eraser"]
items.append("Marker")
print("Problem 4 - After adding:", items)
items.remove("Pencil")
print("Problem 4 - After removing:", items)
```

```python
# 8. Print numbers from 1 to 20 using a loop[cite: 92].
# Hint: Use range(1, 21)[cite: 93].
print("Problem 8 - Numbers 1 to 20:")
for i in range(1, 21):
    print(i, end=" ")
print() # For a new line

# 9. Print only even numbers from 1 to 30[cite: 94].
# Hint: Check if number % 2 == 0[cite: 95].
print("Problem 9 - Even numbers 1 to 30:")
for i in range(1, 31):
    if i % 2 == 0:
        print(i, end=" ")
print()

# 10. Find the sum of numbers from 1 to 50[cite: 96].
# Hint: Start total = 0 and add inside the loop[cite: 97].
total_sum = 0
for i in range(1, 51):
    total_sum += i
print("Problem 10 - Sum of 1 to 50:", total_sum)
```

```python
# 11. greet(name) function[cite: 99].
# Hint: Define and call with two names[cite: 100, 101].
def greet(name):
    print(f"Hello, {name}!")

print("Problem 11:")
greet("Alice")
greet("Bob")

# 12. add_numbers(a, b) function that returns their sum[cite: 102].
# Hint: Use 'return' and print the call[cite: 103, 104].
def add_numbers(a, b):
    return a + b

print("Problem 12 - Sum is:", add_numbers(8, 12))

# 13. square(number) function[cite: 105].
# Hint: Return number * number[cite: 106].
def square(number):
    return number * number

print("Problem 13 - Squares:")
print("Square of 4:", square(4))
print("Square of 7:", square(7))
print("Square of 10:", square(10))
```

```python
# 14. Check positive/negative, if positive check even/odd[cite: 108, 109].
# Hint: Nested if inside the positive block[cite: 110].
num = int(input("Problem 14 - Enter a number: "))
if num >= 0:
    if num % 2 == 0:
        print("Positive and Even")
    else:
        print("Positive and Odd")
else:
    print("Negative")

# 15. Marks pass/fail (>40), nested grade[cite: 111, 112].
# Hint: Nested if-elif-else for A, B, C[cite: 113].
marks = int(input("Problem 15 - Enter marks: "))
if marks >= 40:
    print("Result: Pass")
    if marks >= 80:
        print("Grade: A")
    elif marks >= 60:
        print("Grade: B")
    else:
        print("Grade: C")
else:
    print("Result: Fail")

# 16. Age above 18, nested above 60 check[cite: 114, 115].
# Hint: Outer if > 18, inner if > 60[cite: 116].
age = int(input("Problem 16 - Enter age: "))
if age > 18:
    print("Adult")
    if age > 60:
        print("Senior Citizen")
else:
    print("Not an adult")
```
