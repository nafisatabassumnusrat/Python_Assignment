```python
# --- Problem 1 ---
# Approach: Standard assignment of basic data types, using comma separation inside the print
# function to output them all on a single line.
name = "Alex"
age = 25
height = 1.75
is_student = True
nothing = None
print(name, age, height, is_student, nothing)

# --- Problem 2 ---
# Approach: Using casting functions int() and str() to transform variable types.
# Note that int() truncates the float 3.9 down to 3 instead of rounding.
x = '42'
y = 3.9
z = True

x_int = int(x)
y_int = int(y)
z_str = str(z)

print(x_int, type(x_int))
print(y_int, type(y_int))
print(z_str, type(z_str))

# --- Problem 3 ---
# Approach: Unpack multiple values in one line, then use tuple unpacking (a, c = c, a)
# to elegantly swap values in Python without a temporary storage variable.
a, b, c = 10, 20, 30
print(f"Before swap: a={a}, b={b}, c={c}")
a, c = c, a
print(f"After swap: a={a}, b={b}, c={c}")

# --- Problem 4 ---
# Approach: Reassign the same variable 'x' to different data types to demonstrate
# Python's dynamic typing, checking the type after each assignment.
var_x = 10
print("Int type:", type(var_x))
var_x = "hello"
print("Str type:", type(var_x))
var_x = 3.14
print("Float type:", type(var_x))

# --- Problem 5 ---
# Approach: Testing falsy and truthy values using bool(), and converting booleans to integers.
print(bool(0))        # Output: False (0 is a falsy value)
print(bool(''))       # Output: False (Empty strings are falsy)
print(bool('hello'))  # Output: True (Non-empty strings are truthy)
print(bool(99))       # Output: True (Non-zero integers are truthy)
print(int(True))      # Output: 1 (True maps to integer 1)
print(int(False))     # Output: 0 (False maps to integer 0)
print(float('3.14'))  # Output: 3.14 (String parses cleanly to a floating-point number)
```

```python
# --- Problem 6 ---
# Approach: Using f-strings to format labels alongside mathematical evaluations.
a, b = 15, 4
print(f"Addition: {a + b}")
print(f"Subtraction: {a - b}")
print(f"Multiplication: {a * b}")
print(f"Division: {a / b}")
print(f"Floor division: {a // b}")
print(f"Modulus: {a % b}")
print(f"Exponentiation: {a ** b}")

# --- Problem 7 ---
# Approach: Using augmented operators (e.g., -=, *=) to apply math in-place.
x = 100
x -= 25
print(f"After subtract 25: {x}")
x *= 2
print(f"After multiply by 2: {x}")
x //= 3
print(f"After floor-divide by 3: {x}")
x %= 7
print(f"After modulus 7: {x}")
x **= 2
print(f"After raise to power 2: {x}")

# --- Problem 8 ---
# Approach: Evaluating equality (==) versus identity (is).
a, b = 15, 15
c, d = [1, 2], [1, 2]
print(f"a == b: {a == b}")
print(f"a is b: {a is b}")
print(f"c == d: {c == d}")
print(f"c is d: {c is d}")

# --- Problem 9 ---
# Approach: Using chained comparisons and logical 'or' / 'not' statements.
num1 = 25
string_val = 'yes'
num2 = 7

print(f"10 <= {num1} <= 50: {10 <= num1 <= 50}")
print(f"'{string_val}' is 'yes' or 'no': {string_val == 'yes' or string_val == 'no'}")
print(f"{num2} is NOT divisible by 2: {not (num2 % 2 == 0)}")

# --- Problem 10 ---
# Approach: Using 'in' and 'not in' operators to verify substrings and list items.
sample_list = [1, 3, 5, 7, 9]
print("Is 'py' in 'python'?", 'py' in 'python')
print("Is 'z' not in 'python'?", 'z' not in 'python')
print("Is 7 in list?", 7 in sample_list)
print("Is 4 not in list?", 4 not in sample_list)
```

```python
# --- Problem 11 ---
# Approach: Calling .strip() to clean whitespaces, saving it, and chaining string methods.
s = ' python programming '
s_stripped = s.strip()

print(s_stripped)
print(s_stripped.upper())
print(s_stripped.title())
print(s_stripped.replace('python', 'data'))
print(len(s_stripped))

# --- Problem 12 ---
# Approach: Splitting by comma-space to get clean words, using list indices, and
# applying the string .join() method.
text = 'Data, Science, Machine, Learning'
text_list = text.split(', ')

print("List:", text_list)
print("Second element:", text_list[1])
print("Total elements:", len(text_list))
print("Joined:", '→'.join(text_list))

# --- Problem 13 ---
# Approach: Using string slicing syntax [start:stop:step] to extract substrings.
sentence = 'Hello World Python'

print("First 5:", sentence[:5])
print("Last 6:", sentence[-6:])
print("Every second char:", sentence[::2])
print("Reversed:", sentence[::-1])

# --- Problem 14 ---
# Approach: Directly calling string evaluation methods to locate characters and check digits.
word = 'Banana'

print("find('a'):", word.find('a'))
print("count('a'):", word.count('a'))
print("startswith('Ba'):", word.startswith('Ba'))
print("endswith('na'):", word.endswith('na'))
print("isdigit():", word.isdigit())

print("'12345' isdigit():", '12345'.isdigit())
print("'123a5' isdigit():", '123a5'.isdigit())

# --- Problem 15 ---
# Approach: Formatting floats to two decimal places across f-strings, format(), and % modes.
product = 'Laptop'
price = 75000.5
discount = 10

# 1. f-string style
print(f"Product: {product} | Price: {price:.2f} | Discount: {discount}%")

# 2. .format() style
print("Product: {} | Price: {:.2f} | Discount: {}%".format(product, price, discount))

# 3. % operator style
print("Product: %s | Price: %.2f | Discount: %d%%" % (product, price, discount))
```

```python
# --- Problem 16 ---
# Approach: Ask for user input, cast the birth year string into an integer, and calculate
# against a hardcoded current year (2025).
# Test cases: Try inputting "Alice" and "2000" (outputs 25) vs "Bob" and "1995" (outputs 30).
user_name = input("Enter your name: ")
birth_year = int(input("Enter your birth year: "))
current_year = 2025
age = current_year - birth_year
print(f"Hello {user_name}! You are approximately {age} years old.")

# --- Problem 17 ---
# Approach: Use input() wrapped in float() to do math, outputting the result with :.2f.
# Test cases: Try Length 5, Width 3.5 (Area=17.50, Perim=17.00) vs Length 10.1, Width 2.3.
length = float(input("Enter rectangle length: "))
width = float(input("Enter rectangle width: "))

area = length * width
perimeter = 2 * (length + width)

print(f"Area: {area:.2f}")
print(f"Perimeter: {perimeter:.2f}")
```

```python
# --- Problem 18 ---
# Approach: A top-down if-elif-else ladder to trap the score correctly, followed by a
# secondary independent conditional to determine pass/fail status.
score = int(input("Enter your score (0-100): "))

if score >= 90:
    print("Grade: A+")
elif score >= 80:
    print("Grade: A")
elif score >= 60:
    print("Grade: B")
elif score >= 40:
    print("Grade: C")
else:
    print("Grade: F")

if score >= 40:
    print("Status: Pass")
else:
    print("Status: Fail")

# --- Problem 19 ---
# Approach: Using a nested if block so that the password is only evaluated if the
# username matches the stored credentials.
correct_user = "admin"
correct_pass = "secret123"

in_user = input("Enter username: ")
in_pass = input("Enter password: ")

if in_user == correct_user:
    if in_pass == correct_pass:
        print("Access Granted")
    else:
        print("Wrong Password")
else:
    print("User Not Found")

# --- Problem 20 ---
# Approach: Resolving the sign (pos/neg/zero) in one ternary line, then using standard
# nesting logic to resolve parity only when the number is not equal to zero.
num = float(input("Enter a number: "))

# Ternary expression
sign = "Positive" if num > 0 else ("Negative" if num < 0 else "Zero")
print(f"The number is {sign}")

# Nested check for even/odd (assuming we are checking integers)
if num != 0:
    if num % 2 == 0:
        print("The number is Even")
    else:
        print("The number is Odd")
```
